```python
# 엑셀을 데이터프레임으로 읽기
import pandas as pd
df = pd.read_excel('starbucks_seoul_list.xlsx', header=0)
```

```python
# 서울에 있는 스타벅스의 모든 지점을 지도위에 마커로 출력

starmap = folium.Map(
    location=[37.5665,126.9780],
    zoom_start=11,
    tiles = 'Stamen Terrain'  
)

for idx in df.index:
  lat = df.loc[idx,'위도']
  lng = df.loc[idx,'경도']

  folium.CircleMarker(
      location=[lat, lng],
      fill = True,
      fill_color='green',
      fill_opacity=1,
      color='yellow',
      weight=1).add_to(starmap)

starmap
```

```python
# -----------------------------------------------------------------
# 스타벅스 주소에서 시군구명 단위로 추출
```

```python
sgg_names = []

for addr in df['주소']:
  sgg = addr.split()[1]
  sgg_names.append(sgg)

sgg_names

df['시군구명'] = sgg_names
df
```

```python
# 스타벅스 구별 매장수
star_count = df.pivot_table(index='시군구명',
                            values='매장명',
                            aggfunc='count').rename(columns={'매장명':'매장수'})
star_count
```

```python
# thousands=',' 쉼표 천단위 날리고 가져오기
seoul_sgg = pd.read_excel('seoul_sgg.xlsx', thousands=',')
seoul_sgg
```

```python
seoul_sgg = pd.merge(seoul_sgg,star_count, how='left', on='시군구명')
seoul_sgg
```

```python
# 주민등록인수 십만명당 매장수

seoul_sgg['거주자_십만명당_매장수'] = seoul_sgg['매장수'] / (seoul_sgg['주민등록인구']/100000)
seoul_sgg
```

```python
# 종사자수 십만명당 매장수

seoul_sgg['종사자_십만명당_매장수'] = seoul_sgg['매장수'] / (seoul_sgg['종사자수']/100000)
seoul_sgg
```

```python
import json

sgg_geojson = json.load(open('seoul_sgg.geojson',encoding='utf-8'))
```

```python
geomap = folium.Map(
    location=[37.5665,126.9780],
    zoom_start=11,
    tiles = 'CartoDB dark_matter'  
)
geomap
```

```python
folium.Choropleth(
    geo_data = sgg_geojson,
    data = seoul_sgg,
    coloumns=['시군구명','거주자_십만명당_매장수'],
    key_on='properties.SIG_KOR_NM').add_to(geomap)


geomap
```

```python
geomap2 = folium.Map(
    location=[37.5665,126.9780],
    zoom_start=11,
    tiles = 'CartoDB dark_matter'  
)

folium.Choropleth(
    geo_data = sgg_geojson,
    data = seoul_sgg,
    coloumns=['시군구명','종사자_십만명당_매장수'],
    key_on='properties.SIG_KOR_NM').add_to(geomap2)


geomap2
```

```python
geomap3 = folium.Map(
    location=[37.5665,126.9780],
    zoom_start=11,
    tiles = 'CartoDB dark_matter'  
)

folium.Choropleth(
    geo_data = sgg_geojson,
    data = seoul_sgg,
    coloumns=['시군구명','매장수'],
    key_on='properties.SIG_KOR_NM').add_to(geomap3)


geomap3
```
